# Восстановление золота из руды

Подготовка прототипа модели машинного обучения для «Цифры». 
<br>Компания разрабатывает решения для эффективной работы промышленных предприятий. 

Модель должна ***предсказать коэффициент восстановления золота из золотосодержащей руды.*** Использование данных с параметрами добычи и очистки. 

Модель поможет ***оптимизировать производство, чтобы не запускать предприятие с убыточными характеристиками.***

Расчёт эффективности.
<br>Нужно смоделировать процесс восстановления золота из золотосодержащей руды.
<br>Эффективность обогащения рассчитывается по формуле:



где:
<br>C — доля золота в концентрате после флотации/очистки;
<br>F — доля золота в сырье/концентрате до флотации/очистки;
<br>T — доля золота в отвальных хвостах после флотации/очистки.

Нужно ***спрогнозировать две величины:***
<br>эффективность обогащения чернового концентрата rougher.output.recovery;
<br>эффективность обогащения финального концентрата final.output.recovery.





Описание стадий:
1. Флотация
<br>Во флотационную установку подаётся смесь золотосодержащей руды. После обогащения получается черновой концентрат и «отвальные хвосты», то есть остатки продукта с низкой концентрацией ценных металлов.
На стабильность этого процесса влияет непостоянное и неоптимальное физико-химическое состояние флотационной пульпы (смеси твёрдых частиц и жидкости).
2. Очистка
<br>Черновой концентрат проходит две очистки. На выходе получается финальный концентрат и новые отвальные хвосты.

Описание данных:
<br>Технологический процесс
<br>Rougher feed — исходное сырье
<br>Rougher additions (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant
<br>Xanthate — ксантогенат (промотер, или активатор флотации);
<br>Sulphate — сульфат (на данном производстве сульфид натрия);
<br>Depressant — депрессант (силикат натрия).
<br>Rougher process (англ. «грубый процесс») — флотация
<br>Rougher tails — отвальные хвосты
<br>Float banks — флотационная установка
<br>Cleaner process — очистка
<br>Rougher Au — черновой концентрат золота
<br>Final Au — финальный концентрат золота
<br>Параметры этапов
<br>air amount — объём воздуха
<br>fluid levels — уровень жидкости
<br>feed size — размер гранул сырья
<br>feed rate — скорость подачи

Наименование признаков:
<br>Наименование признаков должно быть такое:[этап].[тип_параметра].[название_параметра]
<br>Пример: rougher.input.feed_ag
<br>Возможные значения для блока [этап]:
<br>rougher — флотация
<br>primary_cleaner — первичная очистка
<br>secondary_cleaner — вторичная очистка
<br>final — финальные характеристики

Возможные значения для блока [тип_параметра]:
<br>input — параметры сырья
<br>output — параметры продукта
<br>state — параметры, характеризующие текущее состояние этапа
<br>calculation — расчётные характеристики

Формула для расчета ***'recovery'*** с условными обозначениями из train:
<br>rougher.output.recovery = (rougher.output.concentrate_au * (rougher.input.feed_au - rougher.output.tail_au)) / (rougher.input.feed_au * (rougher.output.concentrate_au - rougher.output.tail_au))*100

<br>Посчитаем 'rougher.output.recovery' для проверки расчета эффективности обогащения после флотации и сравнение МАЕ по расчитанному показателю и представленному в train.

```python
# Удаление расчитываемого показателя
train_dropna = train.dropna(axis='index', how='any', subset=['rougher.output.recovery'])
```

```python
# Значение rougher.output.recovery по выборке train
train_recovery = train_dropna['rougher.output.recovery']
```

```python
# Расчет rougher.output.recovery
train_recovery_calculation = pd.Series(
    (train_dropna['rougher.output.concentrate_au'] * 
    (train_dropna['rougher.input.feed_au'] - train_dropna['rougher.output.tail_au'])) / 
    (train_dropna['rougher.input.feed_au'] * 
    (train_dropna['rougher.output.concentrate_au'] - 
    train_dropna['rougher.output.tail_au']))*100)
```

```python
MAE = mean_absolute_error(train_recovery, train_recovery_calculation)
print('MAE: {}'.format(MAE))
```

Расчет показателя 'rougher.output.recovery' произведен верно о чем свидетельствует MAE своим малым значением.

Для заполнения пустых значений выборок 'train' и 'test' используем столбец 'date' с перенесением по индексу данных из датасета 'full'.
<br>Для этого добавим столбец 'date_index' во все 3 датасета.

```python
# Цикл для добавления столбца 'date_index'
def index(full):
    full_index= full
    full_index['date_index'] = full['date']
    
    full_index = full_index.set_index('date')
    
    columns = full_index.columns.tolist()
    columns = columns[-1:] + columns[:-1]
    full_index = full_index[columns]
    
    full_index['date_index'] = pd.to_datetime(full_index['date_index'], format='%Y-%m-%dT%H:%M:%S')
    
    return full_index
```

```python
full_index = index(full)
train_index = index(train)
test_index = index(test)

train_index.head()
```

<br>Свинец (Pb)- также прирастает начиная с этапа сырья более чем 2,5 раза;
<br>Вспомогательная добавка (Sol)- показывает динамику снижения, после флотации падает в 3 раза.

##### Вычисление итоговой sMAPE

Сохраняем функцию для вычисления sMAPE и final_sMAPE для проверки лучшей модели на train и test.

```python
def sMAPE(target, predict):
    error = np.mean(abs(target - predict)/((abs(target)+abs(predict))/2))*100
    return error
```

```python
def final_sMAPE(sMAPE_rougher,sMAPE_final):
    return (0.25*sMAPE_rougher + 0.75*sMAPE_final)
```

```python
sMAPE_scorer = make_scorer(sMAPE,greater_is_better=False)
```

Следующий этап работы предполагает кросс-валидацию, выбор лучшей модели на обучающей выборке.

```python
# Создание pipeline для каждой модели
model_lr = make_pipeline(StandardScaler(), SimpleImputer(missing_values=np.nan), LinearRegression())
model_kn = make_pipeline(StandardScaler(), SimpleImputer(missing_values=np.nan), KNeighborsRegressor())
model_rg = make_pipeline(StandardScaler(), SimpleImputer(missing_values=np.nan), Ridge())
```

```python
# Кросс-валидация rougher показателя по моделям
lr_rougher=cross_val_score(model_lr, features_train_rougher, target_train_rougher, scoring=sMAPE_scorer, cv=5).mean().round(2)
print('LinearRegression_rougher: ', lr_rougher)

kn_rougher=cross_val_score(model_kn, features_train_rougher, target_train_rougher, scoring=sMAPE_scorer, cv=5).mean().round(2)
print('KNeighborsRegressor_rougher: ', kn_rougher)

rg_rougher=cross_val_score(model_rg, features_train_rougher, target_train_rougher, scoring=sMAPE_scorer, cv=5).mean().round(2)
print('Ridge_rougher: ', rg_rougher)
```

```python
# Кросс-валидация final показателя по моделям
lr_final=cross_val_score(model_lr, features_train_final, target_train_final, scoring=sMAPE_scorer, cv=5).mean().round(2)
print('LinearRegression_final: ', lr_final)

kn_final=cross_val_score(model_kn, features_train_final, target_train_final, scoring=sMAPE_scorer, cv=5).mean().round(2)
print('KNeighborsRegressor_final: ', kn_final)

rg_final=cross_val_score(model_rg, features_train_final, target_train_final, scoring=sMAPE_scorer, cv=5).mean().round(2)
print('Ridge_final: ', rg_final)
```

```python
# Расчет final_sMAPE по моделям
lr_final_sMAPE=final_sMAPE(lr_rougher,lr_final).round(2)
print('LinearRegression_final_sMAPE: ', lr_final_sMAPE)

kn_final_sMAPE=final_sMAPE(kn_rougher,kn_final).round(2)
print('KNeighborsRegressor_final_sMAPE: ', kn_final_sMAPE)

rg_final_sMAPE=final_sMAPE(rg_rougher,rg_final).round(2)
print('Ridge_final_sMAPE: ', rg_final_sMAPE)
```

Лучшей моделью по метрике sMAPE получилась LinearRegression, проверим ее далее также на тестовой выборке.

```python
# Создание pipeline для LinearRegression на test, расчет final_sMAPE
model_lr_test = make_pipeline(StandardScaler(), SimpleImputer(missing_values=np.nan), LinearRegression())

lr_rougher_test=cross_val_score(model_lr_test, features_test_rougher, target_test_rougher, scoring=sMAPE_scorer, cv=5).mean().round(2)
print('LinearRegression_rougher_test: ', lr_rougher_test)

lr_final_test=cross_val_score(model_lr_test, features_test_final, target_test_final, scoring=sMAPE_scorer, cv=5).mean().round(2)
print('LinearRegression_final_test: ', lr_final_test)

lr_final_sMAPE_test=final_sMAPE(lr_rougher_test,lr_final_test).round(2)
print('LinearRegression_final_sMAPE_test: ', lr_final_sMAPE_test)
```
